2.贴图部分

-纹理映射

这次的纹理映射很简单，只要把每个像素对应的 uv 坐标插值算出来带入就可以了不用mipmap或者双线性插值之类的（mipmap是不是有点像是矩阵求和）

-法线贴图与凹凸贴图

法线贴图和凹凸贴图都可以展现出来平面凹凸的感觉，不过两者的原理不同，简言之，就像法线贴图是改变的表象，而凹凸贴图是改变了本因

凹凸贴图是通过修改法线从而使得模型表面的贴图产生凹凸的错觉，而位移贴图则是真正的修改模型上的顶点，效果会更好但是模型会需要更精细复杂。
